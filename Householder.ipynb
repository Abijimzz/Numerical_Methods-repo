```python
import numpy as np
```

```python
A = np.array([[4,1,-2,2],[1,2,0,1],[-2,0,3,-2],[2,1,-2,-1]])
A
```

```python
def Hh(A,t):
    for i in range(0,t):
        j=i
        n = len(A)
        traza = np.trace(A)
        suma = 0
        
        # G y r
        if (A[i+1,j]<0):
            signo = -1
        else:
            signo = 1
        for it in range(1,n):
            suma += A[it,j]**2
        G = signo*suma**(1/2)
        r = ((1/2)*G**2+(1/2)*A[1,0]*G)**(1/2)
        
        # Vector w
        w = np.zeros((n,n))
        w[0,0]=0
        if i < 1:
            w[1,0]=(A[i+1,0]+G)/(2*r)
        else: 
            w[i,0]=0
        for it in range(2,n):
                w[it,0]=A[it,0]/(2*r)
        
        # Transformacion Hh
        I = np.identity((n))
        P = I - 2*w@np.transpose(w)
        
        # Transformacion de semejanza
        A=P@A@P
        trazai = np.trace(A)
        
        print(f'traza: {traza}\n G:{G} \n r:{r}\n w:{w}\n P{i+1}:{P}\n A{i+2}:{A}\n traza: {trazai}')
```

```python
Hh(A,3)
```
